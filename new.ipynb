```python
import pandas as pd
```

```python
df1 = pd.read_csv("emails.csv")
df2 = pd.read_csv("spam_ham_dataset.csv")
df3 = pd.read_csv("spam_or_not_spam.csv")
```

```python
df1.isnull().sum()
df2.isnull().sum()
df3.isnull().sum()
```

```python
df3.dropna(inplace=True)
```

```python
df2.drop(["Unnamed: 0","label"],axis='columns',inplace=True)
df2
```

```python
df1.rename(columns={"text":"email","spam":"label"},inplace=True)
df2.rename(columns={"text":"email","label_num":"label"},inplace=True)
df = pd.concat([df1,df2,df3])
```

```python
df.label.value_counts()
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(df.email,df.label,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
temp = CountVectorizer()
xtrain_cv = temp.fit_transform(xtrain.values)
```

```python
xtrain_cv.toarray()
```

```python
xtest_cv = temp.transform(xtest.values)
```

```python
temp.get_feature_names_out()
```

```python
dir(temp)
```

```python
temp.vocabulary_
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(xtrain_cv,ytrain)
ypred = model.predict(xtest_cv)
print(classification_report(ytest,ypred))
```

```python
from sklearn.linear_model import SGDClassifier
model_SGD = SGDClassifier()
model_SGD.fit(xtrain_cv,ytrain)
ypred_sgd = model_SGD.predict(xtest_cv)
print(classification_report(ytest,ypred_sgd))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(xtrain_cv.toarray(),ytrain)
ypred = model.predict(xtest_cv.toarray())
print(classification_report(ytest,ypred))
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(xtrain_cv,ytrain)
ypred = model.predict(xtest_cv)
print(classification_report(ytest,ypred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(xtrain_cv,ytrain)
ypred = model.predict(xtest_cv)
print(classification_report(ytest,ypred))
```

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
model = AdaBoostClassifier()
model.fit(xtrain_cv,ytrain)
ypred = model.predict(xtest_cv)
print(classification_report(ytest,ypred))
```

```python
model = RandomForestClassifier()
model.fit(xtrain_cv,ytrain)
ypred = model.predict(xtest_cv)
print(classification_report(ytest,ypred))
```
